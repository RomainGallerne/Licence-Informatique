### Q10-Q11
 **Question 10** Parmis tous les élément de $\mathbb{Z}/17\mathbb{Z}$ (l'ensemble des entiers modulo 17), trouver ceux qui sont des carrés. **Question 11** On se donne la proposition `propSOS` suivante: tout élément de $\mathbb{Z}/n\mathbb{Z}$ est la somme de deux carrés de $\mathbb{Z}/n\mathbb{Z}$. En utilisant la liste `S17` montrer que cette propriété est vrai pour $\mathbb{Z}/17\mathbb{Z}$. Vous affecterez la variable `SOS17` avec le résultat de votre calcul.

```python
# NE PAS MODIFIER
q10=0
# REPONSE QUESTION 10 CELLULE SUIVANTE
```

```python
A17 = Integers(17)
S17 = []
for i in range(0,17):
    for j in range(0,17):
        if(i == j*A17(j) and i not in S17):
            S17.append(i)
print(S17)
```

```python
# NE PAS MODIFIER
q11=0
# REPONSE QUESTION 11 CELLULE SUIVANTE
```

```python
SOS17 = True
for i in range(0,17):
    isomme = False
    for s1 in S17:
        for s2 in S17:
            if(i == A17(s1)+A17(s2)):
                isomme = True
    SOS17 = SOS17 and isomme
print(SOS17)
```
